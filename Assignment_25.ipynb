<b>2) What is the relationship between generators and iterators?

Both generators and iterators use lazy evaluation and use next() to give an output. They may also be used with a loop to get the output

<b>3) What are the signs that a function is a generator function?

- Generator uses yield instead of return
- Generators don't give the output upon calling. They must be iterated using next or a loop

<b>4) What is the purpose of a yield statement?

yield statement converts a function into a generator. yield suspends the function execution and returns a corresponding output and also stores the current state of the generator

<b>5) What is the relationship between map calls and list comprehensions? Make a comparison and
contrast between the two.

- map needs a function to transform the values of an iterable. While list comprehension doesn't need any function to apply a transform to a list.
- Map returns a map object, while list comprehension return a list itself

```python
map_example = map(lambda x: x ** 2, [1, 2, 3, 4, 5])
print([*map_example])
```

```python
list_comp_example = [i ** 2 for i in [1, 2, 3, 4, 5]]
print(list_comp_example)
```
